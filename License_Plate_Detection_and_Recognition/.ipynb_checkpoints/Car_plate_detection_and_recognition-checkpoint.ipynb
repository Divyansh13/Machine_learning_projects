### Load svc model which is used to predict license digit

```python
from sklearn.externals import joblib
digit_model = joblib.load("svc_digit_recognition.joblib")
```

```python
### Create sort_contours() function to grab the contour of each digit
```

```python
def sort_contours(cnts):

    reverse = False
    i = 0
    boundingBoxes = [cv2.boundingRect(c) for c in cnts]
    (cnts, boundingBoxes) = zip(*sorted(zip(cnts, boundingBoxes),
                                        key=lambda b: b[1][i], reverse=reverse))
    print(len(cnts))
    return cnts
```

```python
kernel3 = cv2.getStructuringElement(cv2.MORPH_RECT, (3, 3))
thre_mor = cv2.morphologyEx(binary, cv2.MORPH_DILATE, kernel3)
#cont, _  = cv2.findContours(thre_mor, cv2.RETR_LIST, cv2.CHAIN_APPROX_SIMPLE)
cont, _  = cv2.findContours(thre_mor, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

digit_w = 30
digit_h = 60 

plate_info = ""

# crop 
test_roi = roi.copy()
crop_digits = []

for c in sort_contours(cont):
    (x, y, w, h) = cv2.boundingRect(c)
    ratio = h/w
    if 1<=ratio<=3.5: # Chon cac contour dam bao ve ratio w/h
        if h/roi.shape[0]>=0.5: # Chon cac contour cao tu 60% bien so tro len
            # Draw bounding box arroung digit number
            cv2.rectangle(test_roi, (x, y), (x + w, y + h), (0, 255,0), 2)

            # Sperate number and gibe prediction
            curr_num = thre_mor[y:y+h,x:x+w]
            curr_num = cv2.resize(curr_num, dsize=(digit_w, digit_h))
            _, curr_num = cv2.threshold(curr_num, 220, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)
            crop_digits.append(curr_num)
            #curr_num = np.array(curr_num,dtype=np.float32)
            #curr_num = curr_num.reshape(-1, digit_w * digit_h)

plt.figure(figsize=(10,5))            
plt.imshow(test_roi)
plt.title('test_roi')
#plt.savefig('grab_digit_contour.png',dpi=300)
```

```python
def pre_process(image):
    import numpy as np
    import cv2
    img = np.asarray(image,dtype=np.float32)
    #print(img.shape)
    img = img[np.newaxis,:].reshape(-1,60*30)
    print(img.shape)
    return img
```

```python
fig = plt.figure(figsize=(15,3))
col = len(crop_digits)
row = 1

for digit in crop_digits:
    fig.add_subplot(1,col,row)
    predict_digit = pre_process(digit)
    plt.title('%s'%(digit_model.predict(predict_digit)))
    plt.imshow(digit,cmap='gray')
    row+=1
#plt.savefig('digit_crops.png', dpi=300)
```
